```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
def download_data(ticker,test_data=False):
    '''You can defenitely use a list ,
    but this function is supposed to enter only a single item of your choice'''
    data=yf.download([ticker],auto_adjust=False,start='2025-01-01',end='2026-01-31',period='1d')
    data.index = pd.to_datetime(data.index)
    data_close=data["Adj Close"]
    data_close_main=data_close.loc["2025-01-01":"2025-12-31"]
    data_close_test=data_close.drop(data_close_main.index)
    
    if test_data:
        return data_close_test
    else:
        return data_close_main
```

```python
def export_to_csv(ticker,name,test_data=False):
    data=download_data(ticker,test_data=test_data)
    data.to_csv(f"{name}.csv")
    
```

```python
export_to_csv("RELIANCE.NS","reliance_test",test_data=True)
```
